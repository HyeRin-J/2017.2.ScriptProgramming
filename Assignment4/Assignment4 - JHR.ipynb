```python
import os

#패스워드 암호화 함수
def password_hash(PW):
    import hashlib
    
    hasher = hashlib.sha1()
    PW = PW.encode('utf-8')
    hasher.update(PW) 
    return hasher.hexdigest()

#파일에서 원하는 정보를 불러오는 함수
def readfile(n):
    f = open('Assign2File\\access.txt')
    s = f.read()
    f.close()
    s = s.split()
    List = [s[x] for x in range(len(s)) if x % 4 == n]
    
    for x in range(len(List)):
        List[x] = List[x].replace(':', '')
        List[x] = List[x].replace(',', '')

    return List

#메뉴 함수
def menu():
    print("Welcome to Our Service")
    print("1. Sign Up")
    print("2. Sign In")
    print("3. Quit")

    return int(input('Select menu:'))

while 1:
    sel = menu()
    idList = readfile(0)  #id리스트
    pwList = readfile(1)  #pw리스트
    nameList = readfile(2)  #이름리스트
    print()
    if sel < 0 or sel > 3: #숫자 잘못 입력시
        continue
    if sel == 1: #Sign Up
        ID = input('id: ')
        #id 중복 검사
        while 1:
            if(idList.count(ID) == 0):
                break;
            else:
                print ('Sorry, the entered ID is already used.')
                ID = input('id: ')
        
        PW = input('password: ')
        NA = input('name: ')
        SC = input('school: ')
        PW = password_hash(PW) #패스워드 암호화
        
        #access.txt 파일에 문자열로 저장
        f = open('Assign2File\\access.txt', 'a')
        s = ID + ": " + PW + ", " + NA + ", " + SC + "\n"
        f.write(s)
        f.close()
    if sel == 2:         
        ID = input('id: ')
        
        #id가 안에 있는지 검사
        while 1:
            if(idList.count(ID) == 1):
                idindex = idList.index(ID) #access.txt에서 몇번째 id인지 인덱스 반환
                break;
            else:
                print ('Sorry, you are not a registered member.')
                ID = input('id: ')
                
        PW = input('password: ')
        PW = password_hash(PW) #패스워드 암호화
        #패스워드 검사
        while 1:
            if(PW in pwList[idindex]): #id인덱스에 맞는 패스워드인지 검사
                print('Hello ' + nameList[idindex] + '!') #id인덱스에 등록된 이름 출력
                break;
            else:
                print ('Sorry, the entered password is not correct.')
                PW = input('password: ')
                PW = password_hash(PW)
    if sel == 3: #quit
        break
    print()
```

## 3번 문제
range() 함수와 유사한 frange() 함수를 다음 조건을 참고하여 만드시오.

* 1) frange()의 인자 구성은 다음과 같이 range와 동일하지만 각 인수들은 음수를 받지 않는다고 가정한다.
> range(stop)
> range(start, stop[, step])
* 2) frange() 함수의 인자에 대한 기본 시작(start) 값은 0.0이고, 기본 단계(step) 값은 0.1이다.
* 3) frange 사용 예
    * 3-1) frange(0.5)
        * [0.0, 0.1, 0.2, 0.3, 0.4]
    * 3-2) frange(1.0, 2.0)
        * [1.0, 1.1, 1.2, 1.3, 1.4, 1.5, 1.6, 1.7, 1.8, 1.9]
    * 3-3) frange(2.2, 4.0, 0.5)
        * [2.2, 2.7, 3.2, 3.7]

```python
def frange (start = 0.0, stop = 0.0, step = 0.1):
    import math

    # 인자가 1개일때, start로 들어가는 인자를 stop으로 바꿔줌
    if start > stop:
        start, stop = 0.0, start
    x = []
    # start부터 stop까지 step만큼 더한 리스트를 생성.
    while start < stop:
        #step이 소수점이라 원하는 연산이 안 나와서, 소수점 이하 자리를 버리도록 함.
        x.append(round(start, int(math.ceil(-math.log(step, 10)))))
        start = start + step
    return x

print(frange(0.5))
print(frange(1.0, 2.0))
print(frange(2.2, 4.0, 0.5))
print(frange(2.2, 2.21, 0.001))
```

전부 default 값을 줄 경우, 인자가 1개일 때 값이 start로 들어가게 된다. 이 경우엔 start로 들어가는 값을 stop으로 바꾸고, start값을 다시 0.0으로 초기화하도록 하였다.  
소수점 연산이라 0.1을 더했음에도 1.30000003 같은 결과가 나와서, step의 소수점 자리수 밑으로 반올림한 값을 리스트에 저장하여 반환하도록 하였다.  
* int(math.ceil(-math.log(step, 10))) -> 소수점 자리수를 구함

## 4번 문제
가변인수를 받는 함수 sum()을 다음과 같은 조건을 참고하여 구현하시오

* sum() 사용 예
    * sum()
        * 0
    * sum(1, 2)
        * 3
    * sum(1, 2, 3, 4, 5)
        * 15
    * sum(1, 5, 7, 2, -10)
        * 5

```python
from functools import reduce

def mysum(a = 0, *b):
    b = list(b) #tuple에서 list로 전환
    b.append(a) #list에 a 추가
    return reduce(lambda x, y: x + y, b, 0) #reduce를 사용함, initial = 0

print(mysum())
print(mysum(1, 2))
print(mysum(1, 2, 3, 4, 5))
print(mysum(1, 5, 7, 2, -10))
```

a는 기본인자로 default가 0이고, b는 가변인자이다.  
b가 tuple형태이기 때문에, list형태로 변환해 주고 b에 a를 append한다.  
reduce를 이용하여, 각각 합을 순차적으로 구한 후 반환한다.

여러 단어로 이루어진 문자열을 입력받아 각 단어의 첫글자로 이루어진 단어를 대문자로 출력하는 myinitial() 함수를 다음 조건을 참고하여 작성하시오.

* 1) 다음에 제시되는 함수들을 모두 이용해야 한다.
    * split
    * map
    * join
* 2) myinitial() 함수 사용 예
    * myinitial("as soon as possible")
        * ASAP

```python
def myinitial(string):
    return ''.join(list(map(lambda x: x.capitalize()[0], string.split())))

print(myinitial("as soon as possible"))
```

string을 인자로 받아 공백을 기준으로 split한 리스트의 문자열을 각각 첫글자만 대문자화 후에, 첫번째 글자의 리스트를 map하여 join한다.

## 6번 문제
음이 아닌 정수 n를 입력받으면 n! (factorial)을 계산하는 myfact() 함수를 재귀적 함수로 구현하시오.

* [옵션] 참을 수 있는 정도 만큼의 수행시간을 직접 기다려보면서 n을 늘려보도록 합시다.

```python
def myfact(n):
    if n == 1:
        return 1
    else:
        return n * myfact(n-1)

x = 1950
while x < 2000:
    print(x, ":", myfact(x))
    x = x + 1
```

1900이 넘어가면서부터 시간이 점점 오래걸리기 시작하더니 결국 다음과 같은 에러가 발생했다.  
RecursionError: maximum recursion depth exceeded in comparison  
파이썬 문서를 찾아보니  

> sys.setrecursionlimit(limit)  
Set the maximum depth of the Python interpreter stack to limit. This limit prevents infinite recursion from causing an overflow of the C stack and crashing Python.

> The highest possible limit is platform-dependent. A user may need to set the limit higher when she has a program that requires deep recursion and a platform that supports a higher limit. This should be done with care, because a too-high limit can lead to a crash.

라고 설명하고 있다. limit를 사용자가 설정할 수는 있지만 되도록 반복문을 사용하는 방법을 고려해보아야겠다.

            * 반드시 코딩 내에 similarity 라는 사전 자료 구조가 있어야 함 (즉, 다음과 유사한 코드가 있어야 함).
                * similarity = {}
                * ...
                * similarity[url1] = x1
                * ...
                * similarity[url2] = x2
                * ...
            * 위 코드에서 x1 과 x2는 입력단어와 해당 URL의 웹 사이트 간의 similarity를 나타내는 수치값
        * 유사도가 높은 웹 사이트가 상위에 출력되어야 함 (즉, 유사도 기준 내림 차순)
            * 유사도가 동일한 웹 사이트들에 대해서는 임의 배치함.
* 7) [주의] 필수사항
    * 위에서 만든 검색엔진 코딩은 매우 간단한 것이라 부족한 점이 많이 존재한다.
    * 본인이 생각하기에 상업적인 완성도 높은 검색 로봇/엔진이 되려면 어떤 기능들이 추가적으로 구현되어야 할지 최소 3가지 이상 제시하시오.

```python
def PrintHtml(source):
    # \n, \t 제거
    temp = source.replace('\n', '')
    temp = temp.replace('\t', '')
    temp = temp.replace('\r', '')
    temp = temp.replace('&nbsp', ' ')
    temp = temp.replace('\xa0', ' ')
    temp = temp.replace('&#xb0', '')
    temp = temp.replace('&#039', "'")
    temp = temp.replace('&#8217', "'")
    temp = temp.replace('&#038', '&')
    temp = temp.replace('&#x27', '')

    temp = temp.replace('>', '>$')

    while(temp.find('<script') != -1):
        start =  temp.find('<script')
        end = temp.find('/script>')
        temp = temp[:start] + temp[end+8:]
        
    while(temp.find('<style') != -1):
        start =  temp.find('<style')
        end = temp.find('/style>')
        temp = temp[:start] + temp[end+7:]

    temp = temp.split('<')    
    result = []
    
    x = 0
    
    while x < len(temp):
        z = temp[x]
        if(('src=' in z) or ('href=' in z)):
            temp.remove(z)
        else:
            x = x + 1
    x = 0

    while x < len(temp):
        temp[x] = temp[x].strip()
        if(temp[x].find('$') != -1):
            temp2 = temp[x].split('$')
            y = 0
            while y < len(temp2):
                temp2[y].strip()
                if((temp2[y].find('>') == -1) and (temp2[y] != '')):
                    result.append(temp2[y])
                y = y + 1
        x = x + 1


    result = ' '.join(result)
    result = result.split(' ')

    x = 0
    import string
    while x < len(result):
        for z in string.punctuation:
            result[x] = result[x].replace(z, '')            
        x = x + 1

    d = dict()
    for x in result:
        if(x.isalpha() == True):
            if (x not in d):
                d[x] = 1
            else:
                d[x] = d[x] + 1

    #불용어 제거
    f = open('Assign8File\\stopwords.txt')
    s = f.read()
    s = s.splitlines()
    s.append('—')
    f.close()
    
    for x in s:
        if (d.get(x)):
            d.pop(x)
    
    return (d)
```

script와 style 안의 코드를 삭제하기 위하여 find 메소드를 이용하였다. find에서 반환하는 값이 -1이면 script나 style 태그가 없다는 뜻이기 때문에, 그때까지 반복문을 수행한다. 그 외에 src나 href 태그를 이용하는 문자열은 삭제하도록 하였다.  
불용어를 stopwords라는 파일로 추가하여 파일을 읽어오는 코드를 추가하였다. stopwords.txt에서 읽어온 불용어들을 리스트로 만들고 리스트 내부의 불용어를 pop으로 꺼내면 불용어를 삭제할 수 있다.  
각 단어의 빈도를 저장한 사전인 d를 반환한다.

```python
import pickle

def loadpage(url):
    f = request.urlopen(url)
    source_bytes = f.read()
    source = source_bytes.decode('utf-8')
    filename = url.replace("http://", '')   
    filename = 'Assign8File\\' + filename 

    #html로 저장
    f = open(filename + '.html', 'w', encoding='UTF-8')
    f.write(source)
    f.close()

    d = PrintHtml(source)

    #words frequency 저장
    f2 = open(filename + '.words.frequency', 'wb')
    pickle.dump(d, f2)
    f.close()

    f2 = open(filename + '.words.frequency', 'rb')
    t = pickle.load(f2)
    f2.close()
    
    return t
```

url을 읽어와 html과 각 단어의 빈도를 파일로 저장하는 코드이다.  
단어의 빈도를 리턴받은 d 값을 저장하고, 저장된 words.frequency 파일을 다시 불러와 사전 형태로 받은 후 반환한다.

```python
def wordfrequency(url):
    d = loadpage(url)
    l = list(d.items())
    
    max1 = ('', 0)
    max2 = ('', 0)
    max3 = ('', 0)
    
    for x in l:
        if (x[1] >= max1[1]):
            max3 = max2
            max2 = max1
            max1 = x
        elif(x[1] >= max2[1]):
            max3 = max2
            max2 = x
        elif(x[1] >= max3[1]):
            max3 = x

    result = (url, max1, max2, max3)
    print (result)
    l = dict(l)
    
    return l
```

가장 많이 출연한 단어 3개를 계산하여 출력하는 부분이다. 각각의 url과, 가장 많이 출연한 단어 3개를 출력하도록 하였다.

```python
from urllib import request 

url1 = 'http://www.cnn.com'
url2 = 'http://www.nytimes.com'
url3 = 'http://www.nypost.com'
url4 = 'http://www.foxnews.com'
url5 = 'http://www.latimes.com'

print('가장 많이 출현한 단어 3개')
l1 = wordfrequency(url1)
l2 = wordfrequency(url2)
l3 = wordfrequency(url3)
l4 = wordfrequency(url4)
l5 = wordfrequency(url5)

print()
S = input('Search : ')
similarity = {}

similarity[url1] = l1.get(S, 0)
similarity[url2] = l2.get(S, 0)
similarity[url3] = l3.get(S, 0)
similarity[url4] = l4.get(S, 0)
similarity[url5] = l5.get(S, 0)

similarity = sorted(similarity.items(),key=lambda x:(x[1],x[0]), reverse = True)
result = ''

for x in similarity:
    result = result + x[0] + " "
    
print(result)
```

유저가 입력한 단어와 사이트의 유사도를 저장하는 사전인 similarity를 정의한다. 각 사이트에서 출연한 단어의 빈도를 저장한 사전에서 get메소드를 사용하여 유저가 입력한 단어를 키로 가지고 있는지를 검사하여 그 값을 반환하고, 없으면 0을 반환한다.  
sorted 메소드를 이용하면 사전을 key를 지정하여 정렬할 수 있고, key를 사전의 값을 기준으로 내림차순 정렬하도록 하였다. 만약 값이 같으면 url의 알파벳의 값이 더 큰 url이 먼저 출력된다.
